# 📊 Sales and Demand Forecasting
### Machine Learning Internship Task 1

This project analyzes historical sales data and predicts future demand using a Machine Learning model.

Technologies Used:
- Python
- Pandas
- Matplotlib
- Scikit-learn

```python
# Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## 📂 Load Dataset

In this step we load the sales dataset and preview the data.

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv"

data = pd.read_csv(url)

data.columns = ["Month", "Sales"]

data["Month"] = range(1, len(data)+1)

print("Dataset Preview:")
data.head()
```

## 📊 Dataset Information
Here we analyze the dataset structure and statistics.

```python
print("Dataset Info")
print(data.info())

print("\nStatistics")
print(data.describe())
```

## 📈 Sales Trend Analysis (Line Chart)
This graph shows the sales trend across months.

```python
plt.figure(figsize=(8,5))

plt.plot(data["Month"], data["Sales"], marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

## 📊 Monthly Sales Comparison (Bar Chart)
Bar charts help compare monthly sales values.

```python
plt.figure(figsize=(8,5))

plt.bar(data["Month"], data["Sales"])

plt.title("Monthly Sales Bar Chart")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

## 📉 Moving Average Trend

Moving average helps smooth the sales data and identify trends.

```python
data["Moving_Avg"] = data["Sales"].rolling(window=3).mean()

plt.figure(figsize=(8,5))

plt.plot(data["Month"], data["Sales"], label="Original Sales")
plt.plot(data["Month"], data["Moving_Avg"], label="Moving Average", color="red")

plt.title("Sales Trend with Moving Average")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.legend()

plt.show()
```

## 🤖 Machine Learning Model

We use Linear Regression to predict future sales.

```python
X = data[["Month"]]
y = data["Sales"]
```

```python
model = LinearRegression()

model.fit(X, y)

print("Model Trained Successfully")
```

## 📈 Model Prediction Visualization

```python
predicted_sales = model.predict(X)

plt.figure(figsize=(8,5))

plt.scatter(X, y, label="Actual Sales")

plt.plot(X, predicted_sales, color="red", label="Predicted Trend")

plt.title("Actual vs Predicted Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.legend()

plt.show()
```

## 🔮 Future Sales Forecast
Predicting sales for upcoming months.

```python
future_months = pd.DataFrame({
    "Month":[37,38,39,40,41,42]
})

future_predictions = model.predict(future_months)

forecast = pd.DataFrame({
    "Future_Month":future_months["Month"],
    "Predicted_Sales":future_predictions
})

forecast
```

```python
plt.figure(figsize=(8,5))

plt.plot(data["Month"], data["Sales"], label="Past Sales")

plt.plot(
    future_months["Month"],
    future_predictions,
    marker="o",
    linestyle="dashed",
    label="Forecast"
)

plt.title("Sales Forecasting")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.legend()

plt.show()
```
